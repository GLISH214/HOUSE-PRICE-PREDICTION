```python
#Plotting barplot of property_type column
property_type_variants = d['property_type'].value_counts()
print(property_type_variants )

# Plotting the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=property_type_variants.index, y=property_type_variants.values)
for i, count in enumerate(property_type_variants.values):
    plt.text(i, count + 0.1, str(count), ha='center')
plt.title('Distribution of Property Types')
plt.xlabel('Type')
plt.ylabel('Number')

plt.show()
```

```python

#Top 20 Societies on Barplot


society_types = d['society'].value_counts()
#print(society_types.to_string())

print(society_types)

top_n = 20  # Example: Plotting the top 20 societies

top_societies = society_types.head(top_n)

plt.figure(figsize=(18, 9))
sns.barplot(x=top_societies.index, y=top_societies.values)

plt.xlabel('Society Types')
plt.ylabel('Number')
plt.xticks(rotation=90)

# Annotate each bar with its count
for i, count in enumerate(top_societies.values):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.show()
```

```python

#Removing some societies with less no. of houses/flats using frequency binning to reduce cardinality
# For example, let's say we want to keep societies with at least 10 occurrences
threshold = 10

# Step 3: Filter out societies based on the defined frequency bins
valid_societies = society_types[society_types >= threshold].index

# Filter the DataFrame to keep only rows with societies in valid_societies
d1 = d[d['society'].isin(valid_societies)]
#print(d1.to_string)
d1

# Now, filtered_df contains only rows where the society appears at least threshold times
```

```python

#Plotting the societies after reducing the cardinality

society_counts1 = d1['society'].value_counts()
# print(society_counts1.to_string())
print(society_counts1)



top_n1 = 100  # 
top_societies1 = society_counts1.head(top_n1)

plt.figure(figsize=(40, 20))
sns.barplot(x=top_societies1.index, y=top_societies1.values)
plt.title(f'Top {top_n1} Most Frequent Societies')
plt.xlabel('Society')
plt.ylabel('Count')
plt.xticks(rotation=90)

# Annotate each bar with its count
for i, count in enumerate(top_societies1.values):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.show()
```

```python
#Removing the rows having Independent in the society column

# Filter out rows where 'society' column contains 'independent'
d2 = d1[d1['society'] != 'independent']

# Now, d2 contains only rows where the 'society' column does not contain 'independent'
d2
```

```python

#Price v/s Society
avg_price_by_society = d1.groupby('society')['price'].mean().sort_values()

# Plotting
plt.figure(figsize=(30, 15))
avg_price_by_society.plot(kind='bar', color='skyblue')
plt.title('Average Price by Society')
plt.xlabel('Society')
plt.ylabel('Average Price')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python

missing_values = d2['price'].isnull().sum()
print(missing_values)

print("\nDescriptive Statistics of Price Column:")
print(d2['price'].describe())

plt.figure(figsize=(10, 6))
sns.histplot(d2['price'], kde=True, color='skyblue')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x=d2['price'], color='skyblue')
plt.title('Box Plot of Prices')
plt.xlabel('Price')
plt.show()

skewness = d2['price'].skew()
kurtosis = d2['price'].kurtosis()
print("Skewness:",skewness)
print("Kurtosis", kurtosis)


# Step 6: Price binning
# For example, let's divide prices into three bins: Low, Medium, High
price_bins = pd.qcut(d2['price'], q=4, labels=['Low', 'Medium', 'High','VeryHigh'])
print("\nPrice Bins:")
print(price_bins.value_counts())
```

```python

missing_values1 = d2['price_per_sqft'].isnull().sum()
print(missing_values1)


print(d2['price_per_sqft'].describe())

plt.figure(figsize=(30, 15))
sns.histplot(d2['price_per_sqft'], kde=True, color='skyblue')
plt.title('Distribution of price_per_sqft')
plt.xlabel('price_per_sqft')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(10,5))
sns.boxplot(x=d2['price_per_sqft'], color='skyblue')
plt.title('Box Plot of Prices')
plt.xlabel('price_per_sqft')
plt.show()

skewness = d2['price_per_sqft'].skew()
kurtosis = d2['price_per_sqft'].kurtosis()
print("Skewness:",skewness)
print("Kurtosis", kurtosis)


# Step 6: Price binning
# For example, let's divide prices into three bins: Low, Medium, High
price_per_sqft_bins = pd.qcut(d2['price_per_sqft'], q=4, labels=['Low', 'Medium', 'High','VeryHigh'])
print("\nPrice_per_sqft Bins:")
print(price_per_sqft_bins.value_counts())
```

```python
# List of categorical columns
categorical_columns = ['property_type', 'society','sector','facing','agePossession']  # Add more categorical columns as needed

# List of numerical columns
numerical_columns = ['area', 'bedRoom', 'bathroom','balcony','floorNum','super_built_up_area','built_up_area','carpet_area','study room','servant room','store room','pooja room','furnishing_type','luxury_score']  # Add more numerical columns as needed

# Box plots for categorical columns vs price
for col in categorical_columns:
    plt.figure(figsize=(16,8))
    sns.boxplot(x=col, y='price', data=d2)
    plt.title(f'Price Distribution by {col.capitalize()}')
    plt.xlabel(col.capitalize())
    plt.ylabel('Price')
    plt.xticks(rotation=90)  # Rotate x-axis labels if necessary
    plt.show()

# Scatter plots for numerical columns vs price
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=col, y='price', data=d2)
    plt.title(f'{col.capitalize()} vs Price')
    plt.xlabel(col.capitalize())
    plt.ylabel('Price')
    plt.show()
```

```python
for col in categorical_columns:
    if col in d2.columns:
        g = sns.FacetGrid(d2, col=col, col_wrap=4, height=4, sharex=False)
        g.map(sns.histplot, 'price', bins=30, kde=True)
        g.set_titles(col_template='{col_name}')
        g.set_axis_labels('Price', 'Frequency')
        plt.subplots_adjust(top=0.9)
        g.fig.suptitle(f'Price Distribution by {col.capitalize()}')
        plt.show()
    else:
        print(f"Column '{col}' does not exist in the DataFrame.")
```

```python

d3 = d2.copy()

# Impute missing values with the median
median_price = d3['price'].median()
d3['price'] = d3['price'].fillna(median_price)

print(d3['price'].isnull().sum())
```

```python
print("Null values in price_per_sqft before imputation:", d3['price_per_sqft'].isnull().sum())
```

```python

d4 = d3.copy()


median_price_per_sqft = d4['price_per_sqft'].median()

# Impute missing values with the median
d4['price_per_sqft'] = d4['price_per_sqft'].fillna(median_price_per_sqft)

# Verify that there are no more null values in the price_per_sqft column
print("Null values in price_per_sqft after imputation:", d4['price_per_sqft'].isnull().sum())
```

```python
d5 = d4.copy()


median_area = d5['area'].median()
median_floorNum = d5['floorNum'].median()

# Impute missing values with the median
d5['area'] = d5['area'].fillna(median_area)
d5['floorNum']=d5['floorNum'].fillna(median_floorNum)

# Verify that there are no more null values in the price_per_sqft column
print(d5['area'].isnull().sum())
print(d5['floorNum'].isnull().sum())
```
